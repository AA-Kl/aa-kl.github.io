```python
import os
import shutil

# Define the name of the directory to be created
company_name = "Franklin Templeton"
role_name = "Quantitative Research Analyst"

cv_name = "Arkadiy_Klyukin_CV.tex"
cl_name = "Arkadiy_Klyukin_CL.tex"
cv_source_file = "Template\AA_CV.tex"
cl_source_file = "Template\AA_CL.tex"
vd_source_file = "Vacancy_Description.txt"

# Define the names of the source files and the new files and folder name
source_files = [cv_source_file, cl_source_file, vd_source_file]
new_files = [cv_name, cl_name, vd_source_file]
folder_name = company_name + "_" + role_name

# Check if the folder exists
if not os.path.exists(folder_name):
    os.makedirs(folder_name)
    print(f"Folder '{folder_name}' has been created.")
    # Iterate over the source files and copy their contents to new files
    for source_file, new_file in zip(source_files, new_files):
        source_path = os.path.join(source_file)
        new_path = os.path.join(folder_name, new_file)
        try:
            # Use shutil.copyfile to copy the contents of the file
            shutil.copyfile(source_path, new_path)
            print(f"Contents of '{source_file}' have been copied to '{new_file}'.")
        except FileNotFoundError:
            print(f"The file '{source_file}' does not exist in the folder '{folder_name}'.")
        except Exception as e:
            print(f"An error occurred while copying the files: {e}")
else:
    print(f"The folder '{folder_name}' already exist.")
```
